## train, test 데이터 셋 만들기

```python
from sklearn.model_selection import train_test_split

X_user_top = users_top_age_cheese[['user_age', 'gender', 'buy_count', 'cheese_rank', 'age_group', 'cheese_texture_code', 'brand_code', 'cheese_one_price']]
y_user_top = users_top_age_cheese[['cheese_category_code']]

X_user_train, X_user_test, y_user_train, y_user_test = train_test_split(X_user_top, y_user_top, test_size = 0.3, random_state = 32)

X_user_train.to_csv("resources/data/X_user_top_train.csv", index = False)
X_user_test.to_csv("resources/data/X_user_top_test.csv", index = False)
y_user_train.to_csv("resources/data/y_user_top_train.csv", index = False)
y_user_test.to_csv("resources/data/y_user_top_test.csv", index = False)
```

## 케라스로 모델링 작업

```python
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

def build_model():
    model = Sequential()
    model.add(Dense(8, activation = 'tanh', input_dim=8))
    model.add(Dense(4, activation = 'tanh'))
    model.add(Dense(1, activation = 'softmax'))

    model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics=['accuracy'])
    # model.fit(X_user_train, y_user_train, epochs = 100, batch_size = 5, verbose = 0)
    return model


YourModel = KerasRegressor(build_fn = build_model, epochs = 100, batch_size = 3, verbose = 1)

kf = KFold(n_splits = 5)

results = cross_val_score(YourModel, X_user_train_std, y_user_train, cv = kf)
```
