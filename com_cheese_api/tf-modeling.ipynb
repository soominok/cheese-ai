```python
# 정규화
def min_max_normalize(lst):
    
    normalized = []
    
    for value in lst:
        normalized_num = (value - min(lst)) / (max(lst) - min(lst))
        normalized.append(normalized_num)
        
    return normalized

for x in range(len(X.columns)):
    X_std = min_max_normalize(X[x])

X.describe()
```
